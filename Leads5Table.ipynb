```python
lead_demo = pd.merge(leads, demo, on="lead_id", how="left")

print(lead_demo.head())
```

```python
high_engagement = lead_demo[lead_demo["watched_percentage"] >= 50]

print(high_engagement)
```

```python
lead_calls = pd.merge(leads, calls, on="lead_id", how="left")

print(lead_calls.head())
```

```python
df = leads \
    .merge(demo, on="lead_id", how="left") \
    .merge(calls, on="lead_id", how="left") \
    .merge(reasons, on="lead_id", how="left") \
    .merge(assignment, on="lead_id", how="left")

print(df.head())
```
